```python
from rdflib import Graph, Literal, RDF,DCTERMS, URIRef
from rdflib.namespace import FOAF , XSD
g = Graph()
season = URIRef("https://dbpedia.org/ontology/season")

# Ajouter des triplets à l'aide de la méthode add() de store.
g.add((season, DCTERMS['Devided'], Literal("ETE")))
g.add((season, DCTERMS['Devided'], Literal("AUTOMNE")))
g.add((season,  DCTERMS['Devided'],Literal("HIVER")))
g.add((season, DCTERMS['Devided'], Literal("PRINTEMPS")))
print("--- Mon triplet est composé de : ---")
for s, p, o in g:
    print((s, p, o))
#Voir mon graphe RDF qui est composé en un seul triplet sous format turtle 
print(g.serialize(format='n3').decode("utf-8"))
g.serialize("seaon.rdf", format="pretty-xml")

Humidity = URIRef("http://dbpedia.org/resource/Humidity")

# Ajouter des triplets à l'aide de la méthode add() de store.

g.add((Humidity, DCTERMS['Devided'], Literal("humidity[0,20]")))
g.add((Humidity, DCTERMS['Devided'], Literal("humidity[21,40] ")))
g.add((Humidity, DCTERMS['Devided'], Literal("humidity[41,60]")))
g.add((Humidity, DCTERMS['Devided'], Literal("humidity[41,60]")))
g.add((Humidity, DCTERMS['Devided'], Literal("humidity[81,100]")))


print("--- Mon triplet est composé de : ---")
for s, p, o in g:
    print((s, p, o))
#Voir mon graphe RDF qui est composé en un seul triplet sous format turtle 
print(g.serialize(format='n3').decode("utf-8"))
g.serialize("seaon.rdf", format="pretty-xml")
```

```python
import rdflib
from rdflib.extras.external_graph_libs import rdflib_to_networkx_multidigraph
import networkx as nx
import matplotlib.pyplot as plt
from rdflib import Graph, Literal, RDF, URIRef,DCTERMS
from rdflib.namespace import FOAF , XSD


G = rdflib_to_networkx_multidigraph(g)
```

```python
# Plot Networkx instance of RDF Graph
pos = nx.spring_layout(G, scale=6.6)
edge_labels = nx.get_edge_attributes(G, 'r')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
nx.draw(G, with_labels=True)

#if not in interactive mode for 
plt.show()
```
